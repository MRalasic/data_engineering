```python
try: 
    cur.execute("""CREATE TABLE IF NOT EXISTS song_library (
                song_title text,
                artist_name text,
                year text,
                album_name text,
                single text
               );""")
except psycopg2.Error as e: 
    print("Error: Issue creating table")
    print (e)
```

```python
try: 
    cur.execute("INSERT INTO song_library (song_title, artist_name, year, single, album_name) \
                 VALUES (%s, %s, %s, %s, %s)", \
                 ("Across The Universe", "The Beatles", "1970", "False", "Let It Be"))
except psycopg2.Error as e: 
    print("Error: Inserting Rows")
    print (e)
    
try: 
    cur.execute("INSERT INTO song_library (artist_name, song_title, single, year, album_name) \
                  VALUES (%s, %s, %s, %s, %s)",
                  ("The Beatles", "Think For Yourself", "False", "1965", "Rubber Soul"))
except psycopg2.Error as e: 
    print("Error: Inserting Rows")
    print (e)
```

```python
try: 
    cur.execute("SELECT * FROM song_library;")
except psycopg2.Error as e: 
    print("Error: select *")
    print (e)

row = cur.fetchone()
while row:
   print(row)
   row = cur.fetchone()
```
